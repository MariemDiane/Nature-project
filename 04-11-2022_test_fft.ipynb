```python
from scipy import fft
F2 = fft.rfft2(image, s=None, norm=None)
F2.shape
```

```python
import numpy as np
phase = 2 * np.pi * np.random.rand(512, 257)
```

```python
F_random = np.exp(1j * phase)
Fz_scrambled = F_random * F2
```
